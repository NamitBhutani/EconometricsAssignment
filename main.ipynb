```python
data = pd.read_csv("mysavefile.csv")
data.shape
```

```python
data.head(5)
```

```python
def clean_cols(data):
    data_columns = list(data.columns)
    for i in data_columns:
        if i == 'Unnamed: 0':
            data = data.drop(columns=['Unnamed: 0'])
        else:
            break
    return data

clean_cols(data)
```
